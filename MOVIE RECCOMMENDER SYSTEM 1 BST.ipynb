```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_movies=pd.read_csv('disney movies.csv')
```

```python
df_movies=df_movies.drop(['country', 'date_added'], axis=1)
```

```python
df_movies.columns
```

```python
df_movies.info()
```

```python
df_movies['cast']=df_movies['cast'].fillna(df_movies['cast'].mode() [0])
df_movies['director']=df_movies['director'].fillna(df_movies['director'].mode() [0])
df_movies['rating']=df_movies['rating'].fillna(df_movies['rating'].mode() [0])
```

```python
df_movies.isnull().sum()
```

```python
df_movies.head(40)
```

```python
df_movies=df_movies.rename(columns={'listed_in': 'genre'})
```

```python
df_movies.columns
```

```python
#process the genre column
vectorizer=CountVectorizer(tokenizer=lambda x: x.split(' '))
genre_matrix=vectorizer.fit_transform(df_movies['genre'])

print(vectorizer.get_feature_names_out())
print(genre_matrix.toarray())
```

```python
#compute similarity
similarity_matrix=cosine_similarity(genre_matrix)


#Display similarity matrix
print(pd.DataFrame(similarity_matrix, index=df_movies['title'], columns=df_movies['title']))
```

```python
#make reccommendations
# Function to get top recommendations
def recommend_movie(df_movies_title, num_recommendations=2):
    # Find the index of the movie
    idx = df_movies[df_movies['title'] == df_movies_title].index[0]
    
    # Get similarity scores
    similarity_scores = list(enumerate(similarity_matrix[idx]))
    
    # Sort movies by similarity score
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    # Get the top recommendations
    recommended_indices = [i[0] for i in sorted_scores[1:num_recommendations+1]]
    return df_movies['title'].iloc[recommended_indices]

#get the reccommendation to the specific movie
print("Reccommendation for Ernest Saves Christmas")
print(recommend_movie('Ernest Saves Christmas') ) 


```
